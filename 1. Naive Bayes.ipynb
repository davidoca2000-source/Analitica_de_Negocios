#**Caso de estudio**
En este documento se desarrolla y analiza un modelo Naive Bayes para la PreAprobación de créditos en una entidad finannciera utilizando las variables socio económoicas de un solicitante de crédito.
De acuerdo con lo anterior, las variables socioeconómicas son:

**Edad**: Es la edad que posee el solicitante al momento de solicitar crédito.

**Ingresos**: Ingresos recibidos por una persona como resultado de una actividad comercial.

**Egresos**: Son los gastos mensuales del solicitante. Gastos de personas a cargo, hijos y seervicios públicos.

**Monto (EAD)**: Indica el dinero que una persona solicita a una entidad financiera como crédito.

0. Se procede con la carga de las librerías de trabajo con IA.


```python
import numpy as np #Librería numérica por excelencia.
import pandas as pd #Librería para la comunicación con archivos de Excel.

#Vamos a trabajar con Librerías específicas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos BaseAAA

```python
nxl='/content/1. SolicitantesCrédito(USD) Base.xlsx' #Copiamos la ruta del archivo de Excel
XDB=pd.read_excel(nxl,sheet_name=1)
XDB.head() #Muestreme los datos iniciales de toda la base de datos cargada

#Seleccionamos variables de trabajo
XD=XDB[["Edad", "Ingresos", "Egresos", "Monto (EAD)"]] #Variables de Trabajo
XD.head(10) #La cantidad de filas filas que quiero ver
yd=XDB[["PreApr"]] #Esta es la variable de decisión
yd.head()
```

2. Implementamos Modelo Navie Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd) #Ajustar Variables Entrada - Salida

#Mostramos las medias de las variables
u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("Edad","Ingresos","Egresos","Monto (EAD)")
print(u)
print("Las desviaciones son:")
print(sigma)
```

**Análisis de Resultados**

De los resultados arrojados por el modelo, podemos observar que la categoría de PreAprobados posee ingresos superiores a las categorías de PreNegados, esto indica claramente la predominancia de la variable para aprobar créditos. Es importante mencionar, que la edad es mayor para los PreAprobados que para los PreNegados, lo que indica que el banco busca personas con mayor experiencia crediticia. Podemos decir igualmente, que los PreAprobados tienen ingresos superiores a los de la categoría PreNegados, sin embargo, los PreAprobados están solicitando montos superiores a los de la categoría de PreNegados, esto e sun indicador de la capacidads de este grupo para hacer frente a un crédito.

3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión

```python
ydp=mnb.predict(XD) #Predigame cosas a partir de la matriz XD. Esto es lo que el modelo aprende ydp de pronóstico
cm=confusion_matrix(yd,ydp)
print(cm)

#Se determinan las métricas de la matriz de confusión
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1];TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo. Apartir del 75% es un buen modelo
Ex=(VP+VN)/TDatos
print("Exactitud: ", Ex)

#2. Tasa Error: Porcentaje de fallos del modelo
TEr=(FP+FN)/TDatos
print("Tasa Error: ", TEr)

#3. Sensibilidad: Cómo se comportó con respecto a los positivos solamemte
Se=VP/(VP+FN)
print("Sensibilidad: ",Se)

#4. Especificidad: Cómo se comporta pronosticando negativos
Es=VN/(VN+FP)
print("Especificidad: ",Es)

#5. Precisión: Es una  versión de cómo se comporta el modelo frente a los positivos solamente
Pr=VP/(VP+FP)
print("Precisión: ",Pr)

#6. predicción negativa: Cómo funciona el modelo pronbosticando créditos PreNegados
PrN=VN/(VN+FN)
print("Predicción Negativa: ",PrN)

```

**Análisis de Resultados**

De acuerdo con los resultados, podemos observar de manera general que el modelo alcanzó una exactitud por encima del 75% en promedio, lo que indica el buen comportamiento del modelo frente a la clasificación de los solicitantes de crédito frente a la PreAprobación y la PreNegación. Se destaca la Precisión con un 80.9%, lo que indica el buen funcionamiento del modelo frente a la PreAprobación de créditos. Igualmente, se puede observar el buen comportamiento del modelo frente a la PreNegación de créditos, tal y como lo demuestra el índice de Especificidad del modelo, el cual estuvo por encima del 84% en promedio. Es importante mencional que la exactitud queda por debajo del 80%, ya que esta métrica se ve afectada por los errores en la clasificación de solicitantes de crédito, lo que se evidencia a través de la Sensibilidad y la Predicción Negativa.
